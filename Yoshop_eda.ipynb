```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

## 1. Rating given by customers

```python
df1 = pd.read_csv('review_dataset.csv')
df1.sample(5)
df1.dropna()
df1['stars'] = df1['stars'].replace(['4.9 star rating', '5.0 star rating','4.8 star rating','4.7 star rating', '4.6 star rating'], '>4.5')
df1['stars'] = df1['stars'].replace(['4.2 star rating', '4.3 star rating','4.5 star rating', '4.0 star rating'], '4.0-4.5' )
df1['stars'] = df1['stars'].replace(['3.3 star rating', '2.3 star rating', '3.0 star rating'], '<4')
labels = '>4.5 ','4.0-4.5 ','<4 '
counts = df1['stars'].value_counts()
```

## 6. Reviews for All Product Categories

```python
df3 = pd.read_csv('review_dataset.csv')
Data6 = df3.groupby(['stars','category'])
stars_numpy = df3['stars'].to_numpy()
stars = []
for i in range(len(stars_numpy)):
    stars.append(float(str(stars_numpy[i])[0:3]))
data6 = pd.DataFrame(stars, columns =['star rating'])
df3['stars'] = data6['star rating']
Data6_ = df3.groupby(['category'])[['stars']].mean()
```

## 7. Number of Orders Per Month Per Year

```python
x = 'Order Date and Time Stamp'
date_time_df = df2[x]
date_time_numpy = date_time_df.to_numpy()
per_year_list = []
per_month_list = []
for i in range(2297):
    per_year_list.append(date_time_numpy[i][6:10])
    per_month_list.append(date_time_numpy[i][3:5])
per_year_df = pd.DataFrame(per_year_list, columns = ['Year'])
per_month_df = pd.DataFrame(per_month_list, columns = ['Month'])
```

## 9.Analysis of Number of Orders Across Parts of a Day

```python
date_time_df = df2[x]
date_time_numpy = date_time_df.to_numpy()
hour_list = []
for i in range(2297):
    hour_list.append(date_time_numpy[i][11:13])
hour_list_df = pd.DataFrame(hour_list, columns = ['Time of the day'])
def handlehrs(val):
    if (5 <= int(val) < 12):
        return 'morning'
    if (12 <= int(val) < 17):
        return 'after noon'
    if (17 <= int(val) < 21):
        return 'night'
Time_of_the_day = hour_list_df['Time of the day'].apply(handlehrs)
```
